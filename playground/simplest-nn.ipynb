```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
EPOCHS = 500
```

```python
# Y = 2x â€“ 1
def f(_x: float):
  return 2 * _x - 1
```

```python
x = [-1.0, 0.0, 1.0, 2.0, 3.0, 4.0]
y = [-3.0, -1.0, 1.0, 3.0, 5.0, 7.0]
```

```python
layers = [
  Dense(units=1, input_shape=[1])
]
model = Sequential(layers)
model.compile(optimizer='sgd', loss='mean_squared_error')
model.summary()
```

```python
xs = np.array(x, dtype=float)
ys = np.array(y, dtype=float)
(xs.shape, ys.shape)
```

```python
model.fit(xs, ys, epochs=EPOCHS)
```

```python
_input = 10.
prediction = model.predict([_input])
manual_prediction = f(_input)
y1 = prediction[0][0]
y2 = manual_prediction
print(y1, y2)
```

```python
close_prediction = np.isclose(y1, y2, rtol=0.05)
print(f'Is the prediction close to the known answer? {"Yes" if close_prediction else "No"}')
```
